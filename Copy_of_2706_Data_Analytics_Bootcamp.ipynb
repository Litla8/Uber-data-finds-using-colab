## The Story: Monday Morning at Uber HQ

> *You are a Data Scientist at Uber. Your Product Manager walks in on a Monday morning and says:*
>
> **"Riders are furious! They book a cab, but either the driver cancels at the last minute OR the app says 'No Cars Available'. We are losing thousands of rides per week — especially to and from the airport.**
>
> **Your job: Use data to find out WHY this is happening and WHAT we can do about it."**

---

### 🤔 Think Before You Code

> **Before looking at any data, what questions would YOU ask?** A great analyst always clarifies before diving in.

| ❓ Question to Ask | 🎯 Why It Matters |
|--------------------|------------------|
| Is this happening at specific times of day? | Narrow it down to rush-hour vs. off-peak |
| Is it City → Airport, Airport → City, or both? | Different routes may have different root causes |
| Is the increase sudden or gradual? | Sudden = one event; Gradual = a trend |
| Have driver incentive programs changed recently? | Removing bonuses can trigger mass cancellations |
| Is the failure 'Cancelled' or 'No Cars Available'? | Different failure modes = different solutions |

> 💡 **The data we have can answer most of these questions. That's exactly what we'll do today.**

## Why Do We Import Libraries?

Python by itself is like a **bare kitchen** — it has a stove and a sink, but no pots, pans, or knives. Libraries are the specialist tools we bring in.

**Importing** a library = picking up that tool from the shelf and placing it on your workbench.

### The Syntax
```python
import  <library name>  as  <short nickname>
  ↑                      ↑
"bring in"               "I'll call it this from now on"
```

> 💡 **Why short nicknames?** Without them, you'd write `numpy.array([1,2,3])` every single time. With them: `np.array([1,2,3])`. Saves thousands of keystrokes!

### 🧠 Memory Trick

| Shortcut | How to Remember |
|----------|-----------------|
| `np` | **N**um**P**y — just the first two consonants |
| `pd` | **P**an**D**as — P and D |
| `plt` | **PLo**T |
| `sns` | **S**eabor**N** — named after *Samuel Norman Seaborn*, a TV character the creator liked! |

## What is NumPy and Why Does It Exist?

**NumPy = Numerical Python.** It is the foundation of ALL data science in Python. Pandas, Matplotlib, and Seaborn are all built on top of it.

Think of it as: **a calculator that can handle ONE MILLION numbers in a single shot.**

### Python List vs NumPy Array

```
Python List  [ 1, 2, 3, 4, 5 ]
─────────────────────────────────────────────────
Memory: [ box ] [ box ] [ box ] [ box ] [ box ]
         ↑        ↑       ↑      ↑       ↑
         1        2       3      4       5
Each 'box' has extra overhead. Spread across RAM. SLOW for maths.

NumPy Array  np.array([ 1, 2, 3, 4, 5 ])
─────────────────────────────────────────────────
Memory: | 1 | 2 | 3 | 4 | 5 |  ← like a ruler, side-by-side
No overhead. FAST!  (50–300x faster than Python lists!)

Real world: Processing 1M Uber rides → NumPy: 0.01s  vs  List: 3s
```

### NumPy vs Excel Functions

| NumPy Function | What It Does | Excel Equivalent |
|----------------|-------------|------------------|
| `np.mean(arr)` | Average of all values | `=AVERAGE()` |
| `np.max(arr)` | Largest value | `=MAX()` |
| `np.min(arr)` | Smallest value | `=MIN()` |
| `np.median(arr)` | Middle value when sorted | `=MEDIAN()` |
| `np.std(arr)` | How spread out values are | `=STDEV()` |
| `np.sum(arr)` | Total of all values | `=SUM()` |

```
┌─────────────────────────────────────────────────────────────┐
│  A DataFrame is a TABLE with rows and columns               │
│                                                             │
│   ID    │ Pickup   │ Driver  │ Status                       │
```

## What is Feature Engineering?

Raw data often doesn't have **everything** we need for analysis. We **create** new columns from existing ones.

> **Analogy:** You have a student's date of birth. From ONE column, you can create:
> - **Age** (right now)
> - **Age group** (Child / Teen / Adult / Senior)
> - **Birth month** (seasonal pattern?)
>
> That's Feature Engineering!

## What We'll Create

| New Column | Source Column | Purpose |
|------------|--------------|---------|
| `RequestHour` | `Request timestamp` | Find rush-hour patterns |
| `TimeSlot` | `RequestHour` | Group hours into named periods |
| `Cab Availability` | `Status` | Simple Available/Not Available flag |
| `Duration_mins` | Both timestamps | Ride duration in minutes |


````python
df['RequestHour']= df['Request timestamp'].dt.hour
````

````python
df.head()
````

````python
df['Status'].value_counts(normalize=True)
````

````python
plt.figure(figsize=(10,6))
sns.countplot(data =df, x='Pickup point', hue='Status')
````

````python
sns.countplot(data =df, x='Pickup point')
````

````python
hourly_demand =df.groupby('RequestHour').size()
````

````python
plt.figure(figsize=(10,6))
sns.lineplot(x=hourly_demand.index, y=hourly_demand.values)
plt.xticks(range(0, 24));
````

````python
hourly_status = df.groupby(['RequestHour', 'Status']).size().unstack()
````

````python
hourly_status.plot(kind='bar', stacked=True, figsize=(14, 6),
                   color=['#F44336', '#FF9800', '#06C167'])
plt.xticks(range(0, 24));
````

````python
df_city = df[df['Pickup point']== 'City']
````

````python
df_city.head()
````

````python
hourly_status = df_city.groupby(['RequestHour', 'Status']).size().unstack()
````

````python
hourly_status.plot(kind='bar', stacked=True, figsize=(14, 6),
                   color=['#F44336', '#FF9800', '#06C167'])
plt.xticks(range(0, 24));
````

````python
df_airport = df[df['Pickup point']== 'Airport']
````

````python
hourly_status = df_airport.groupby(['RequestHour', 'Status']).size().unstack()
````
